```python
X = df_all.drop(['T-P'], axis=1)
y = df_all['T-P']
```

# **Model Selection**

## 1. Multiple Linear Regression

```python
linreg = LinearRegression()

linreg.fit(X_train_Encode, y_train)
y_pred_l = linreg.predict(X_test_Encode)
ldf = pd.DataFrame({'True values' : y_test, 'Predicted values': y_pred_l})
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_l))
mse = mean_squared_error(y_test,y_pred_l)
mae = mean_absolute_error(y_test, y_pred_l)
r2_1 = r2_score(y_test,y_pred_l)
print('RMSE is : {}'.format(rmse_test))
print('MSE is : {}'.format(mse))
print('MAE is : {}'.format(mae))
print('r2 score is : {}'.format(r2_1))
Accuracy1=r2_score(y_test,y_pred_l)*100
print(" Accuracy of the model is %.2f" %Accuracy1)
ldf.head()
```

## 2. RandomForest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators =1000,  max_features = 'sqrt', max_depth = 5, random_state = 42).fit(X_train_Encode, y_train)
y_pred_r = rf.predict(X_test_Encode)
rdf = pd.DataFrame({'True values' : y_test, 'Predicted values' : y_pred_r})
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_r))
mse = mean_squared_error(y_test,y_pred_r)
mae = mean_absolute_error(y_test, y_pred_r)
r2_2 = r2_score(y_test,y_pred_r)
print('RMSE is : {}'.format(rmse_test))
print('MSE is : {}'.format(mse))
print('MAE is : {}'.format(mae))
print('r2 score is : {}'.format(r2_2))
Accuracy2=r2_score(y_test,y_pred_r)*100
print(" Accuracy of the model is %.2f" %Accuracy2)
rdf.head()
```

# **Hyperparameters Tuning**

```python
from scipy.stats import randint, uniform
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
import time

#  fit
pipe = make_pipeline(
    OrdinalEncoder(), 
    XGBRegressor(random_state=2)
)

params = {
    'n_estimators':[1,100],
    'min_child_weight':[0,5], 
    'gamma':[i/10.0 for i in range(3,6)],  
    'subsample':[i/10.0 for i in range(6,11)],
    'colsample_bytree':[i/10.0 for i in range(6,11)], 
    'max_depth': [2,3,4,6,7],
    'objective': ['reg:squarederror', 'reg:tweedie'],
    'booster': ['gbtree', 'gblinear'],
    'eval_metric': ['rmse'],
    'eta': [i/10.0 for i in range(3,6)]
}

reg = XGBRegressor(nthread=-1)

# run randomized search
n_iter_search = 100
random_search = RandomizedSearchCV(reg, param_distributions=params,
                                   n_iter=n_iter_search, cv=5, scoring='neg_mean_squared_error')

start = time.time()
random_search.fit(X_train_Encode, y_train)
print("RandomizedSearchCV took %.2f seconds for %d candidates"
      " parameter settings." % ((time.time() - start), n_iter_search))
```

```python
best_regressor = random_search.best_estimator_
print(best_regressor)
```

```python
y_best = best_regressor.predict(X_train_Encode)
rmse_pred = mean_absolute_error(y_train, y_best) 

print("Root Mean Absolute Error:" , np.sqrt(rmse_pred))
```

# **Partial Dependence Plot(PDP)**

```python
from pdpbox.pdp import pdp_isolate, pdp_plot
from pdpbox.pdp import pdp_interact, pdp_interact_plot
from pdpbox import pdp
import shap
```

```python
def draw_pdp_plot(feature, model, X_val, num_grid_points = 10):
    isolated = pdp_isolate(
        model=model,
        dataset=X_val,
        model_features=X_val.columns,
        feature=feature,
        grid_type='percentile',
        num_grid_points=num_grid_points
    )

    pdp.pdp_plot(isolated, feature)
```

```python
features = X_train_Encode.columns.tolist()[1:]
features
```

```python
X = pd.concat((X_train_Encode, X_test_Encode))

for i in range(8): draw_pdp_plot(features[i], random_search, X)
```
